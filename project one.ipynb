```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data_set=pd.read_csv('911.csv')
data_set.head()
```

```python
data_set['zip'].value_counts().head()
```

```python
data_set['twp'].value_counts().head(7)
```

```python
data_set['title'].nunique()
```

```python
data_set['specific reasons'] = data_set['title'].apply(lambda title: title.split(':')[0])
```

```python
data_set.head()
```

```python
sns.countplot(x='specific reasons',data=data_set)
```

```python
data_set['timeStamp']=pd.to_datetime(data_set['timeStamp'])
data_set.head()
```

```python
data_set['Hour']=data_set['timeStamp'].apply(lambda time: time.hour)
data_set['Minutes']=data_set['timeStamp'].apply(lambda time: time.minute)
data_set['Month']=data_set['timeStamp'].apply(lambda time : time.month)
data_set['Number of Week']=data_set['timeStamp'].apply(lambda time : time.dayofweek)
data_set.head()
```

```python
sns.countplot(x='Month',data=data_set,hue='specific reasons')
```

```python
by_month=data_set.groupby('Month').count()
by_month['specific reasons'].plot()
```
